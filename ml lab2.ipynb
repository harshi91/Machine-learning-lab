```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
```

```python
data=fetch_california_housing()
```

```python
data
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df['Target']=data.target
```

```python
variable_meaning={
    "MedInc":"Median income in block group",
    "HouseAge":"Median house age in block group",
    "AveRooms":"Average number of rooms per household",
    "AveBedrooms":"Average number of bedrooms per household",
    "Population":"Population of block group",
    "Ave0ccup":"Average number of household members",
    "Latitude":"Latitude of block group",
    "longitude":"Longitude of block group",
    "Target":"Median house value(in $100,000s)"
}
```

```python
variable_df=pd.DataFrame(list(variable_meaning.items()),
columns=["Features","Description"])
print("\nVariable Meaning table:")
print(variable_df)
```

```python
plt.figure(figsize=(12,8))
df.hist(figsize=(12,8), bins=30,edgecolor='black')
plt.suptitle("Feature Distributions",fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title=("Boxplot of features to Identity Outliers")
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr_matrix=df.corr()
sns.heatmap(corr_matrix,annot=True, cmap='coolwarm',fmt='.2f')
plt.title("Feature correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['MedInc','HouseAge','AveRooms','Target']],
diag_kind='kde')
plt.show()
```
